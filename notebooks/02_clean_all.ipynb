# Data Cleaning

```python
import gzip
import json

import polars as pl
```

## Building Sets and Membership

Mostly just renaming columns and dropping uninformative ones

```python
# Drop/rename
building_sets = (
    pl.read_csv("../data/raw/building_sets.csv")
    .drop("bsDescription", "bsPublic", "owner")
    .rename({"bsID": "bs_id", "bsName": "name"})
)
set_membership = (
    pl.read_csv("../data/raw/set_membership.csv")
    .rename({"facid": "fac_id", "bsID": "bs_id"})
)

# Save
building_sets.write_csv("../data/clean/building_sets.csv")
set_membership.write_csv("../data/clean/set_membership.csv")
```
